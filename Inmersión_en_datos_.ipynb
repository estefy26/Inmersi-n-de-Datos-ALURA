```python
# PROMEDIO DE INMUEBLES AGRUPADOS POR BARRIO,TENIENDO EN CUENTA EL PRECIO POR M2
inmuebles.groupby('Barrio').mean()
```

```python
# PRIMERO SE REALIZA UNA SUMATORIA POR BARRIO CREANDO UNA NUEVA VARIABLE LLAMDA DATOS_BARRIO PARA QUE NOS 
#DEJE EL PROMEDIO DE PRECIO POR M2 DE CADA UNO COMO VALOR ESPECIFICO ESTE ES EL PROMEDIO REAL DE CADA UNO
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
# AHORA SI DETERMINAMOS EL PROMEDIO DE VALOR POR M2 POR BARRIO REALIZANDO LA OPERACION 
datos_barrio['valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
 # CREAMOS UN DICCIONARIO DENTRO DE LA VARIABLE M2 BARRIO LA CULA NOS INDICA EL M2 ,TOMA COMO LLAVE O KEY EL BARRIO Y EL VALOR EL PRECIO POR M2

 m2_barrio = dict(datos_barrio['valor_m2_Barrio'])
     
```

```python
# LA IDEA ES LLEVAR ESTE NUEVPO VALOR POR M2 A NUESTRO DATA SET INMUEBLES PARA TENR LA REFERENCIA
#PARA TODOS LOS INMUEBLES QUE TIENE EL BARRIO
#SE SUSTITUYE EL VALOR ANTIGUO Y TRAER EL NUEVO VALOR LLAMADO VALOR M2 BARRIO PARA TENER EL NUEVO VALOR POR M2 DE CADA UNO DE LOS BARRIOS
inmuebles['valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['valor_m2_Barrio'] = inmuebles['valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
# REVISAMOS TOP 10 DE CADA UNO DE LOS BARRIOS
inmuebles['Barrio'].value_counts()
```

```python
#revisamos los 10 barrios con mas inmuebles a la venta agregado :10 index que es el indice de los barrios 
inmuebles['Barrio'].value_counts()[:10].index
```

```python
#con lo anterior yo puedo almacenar los abrrios con mas inmuebles a la venta el top 10 en una variable llamada top_barrios
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```

```python
# se convierte el indice de barrios en una columna de la siguiente manera y asi ya queda en nuestar data
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
# se realiza consulta a este data set
# se toma datos barrios y funcion query 
# los barrios que esten en los top la arroba nos arroja la tabla de esa información
datos_barrio.query('Barrio in @top_barrios')
```

```python
# se pueden tomar para un grafico y hallar el grafico mas indicado para represenatr estos dats
# primero el lienzo de 10 por 8
#primero unas barras para luego avanzar con el grafico ideal oara estos datos
#se le atribuye a una variable llamada ax,y se grafica con el top barrios
#
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y="valor_m2_Barrio", data = datos_barrio.query('Barrio in @top_barrios'))
```

```python
#para mejorar le gráfico podemos agregar en nuestra variable ax, los parametros de los tick para ajustarlos aqui es donde estan
#los indicadores cada punto que se queire indicar
# se especifca el eje x y una rotación 
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y="valor_m2_Barrio", data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
#se puede concluir que santa barbara que esta ubicado en la zona norte de bogota, es el mas alto dentro dl data de barrios con mas inmuebles a la venta 
#no necesariamente es el barrio con mayor precio cuadrado, dentro de los top 10 de los barrios con mas inmuebles es el que tiene el precio mayor por m2 solo en nuestar data de los que tienen
#mas inmuebles a la venta no en la data general
```

```python
# copiamos el codigo y se cambia a boxplot es grafico de cajas este si muestra mejor este tipo de dtaos
#es necesario colocar la clausula show para que muestre nuestro grafico como tal
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="precio_por_metro_cuadrado", data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

nos muestra el valor min y max a cada uno de los extremos de la grafica, la linea inferior de cada caja indica el 25% (primer cuartil) de los datos estan por debajo de este valor, ejemplo en barrios unidos esta por 4 millones mas o menos.En la mitad se encuentra la mediana en la mitad de la cajita de cada color entonces el 50%(segundo cuartil) de mi data se encuentra por denbajo de la medina, el tercr cuartil es el 75% de los valores por m2 se encuentran por debajo de ese valor  que seria el cuadrito de arriba y en la parte superior es el valor max, no toma el valor max absoluta si no que excluye lod datos discrepantes, los errores, o fuera de lo normal.

```python
# en este caso puedo expandir mi grafico y en mi propia query colocar un filtro añadiendo una condicion
#en los top barrios el valor del m2 sea hasta los 15 millones

 plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="precio_por_metro_cuadrado", data = inmuebles.query('Barrio in @top_barrios & precio_por_metro_cuadrado < 15'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

asi se puede ver mejor las cajas, valores max y min

```python
# podemos ver otros atributos como area y se condiciona menor a 500
#asi es mejor la visualizacion de los datos sin el filtro se grrafca todas las reas y no se entiende es la idea del filtro
#eliminar los datos discrepantes o outlayers, ya que perjudican nuestro analisis
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Area", data = inmuebles.query('Barrio in @top_barrios & Area < 500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
     
```
